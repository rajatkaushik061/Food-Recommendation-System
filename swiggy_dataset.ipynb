```python
food_list = df['Food type'].dropna().str.split(',')
flat_list = [item.strip() for sublist in food_list for item in sublist]
```

```python
from collections import Counter

food_counter = Counter(flat_list)
top_10_foods = food_counter.most_common(10)
```

```python
labels, values = zip(*top_10_foods)
plt.figure(figsize=(10, 5))
sns.barplot(x=list(values), y=list(labels), palette='viridis')
plt.title('Top 10 Most Common Food Types')
plt.xlabel('Number of Restaurants')
plt.ylabel('Food Type')
plt.tight_layout()
plt.show()
```

```python
def recommend_restaurants_by_area(area_name, df=df, top_n=5):
    area_df = df[df['Area'].str.lower() == area_name.lower()]
    
    if area_df.empty:
        print(f"Sorry, no restaurants found in '{area_name}'.")
        return
    
    # Sort by Avg ratings (descending)
    top_restaurants = area_df.sort_values(by='Avg ratings', ascending=False).head(top_n)

    # Show results
    print(f"\n📍 Top {top_n} Restaurants in {area_name.title()}:")
    print(top_restaurants[['Restaurant', 'Avg ratings', 'Price', 'Food type', 'Delivery time']])
    
    # Plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Avg ratings', y='Restaurant', data=top_restaurants, palette='magma')
    plt.title(f'Top {top_n} Restaurants in {area_name.title()} by Rating')
    plt.xlabel('Average Rating')
    plt.ylabel('Restaurant')
    plt.tight_layout()
    plt.show()

# Example usage:
recommend_restaurants_by_area('Indiranagar')
```
